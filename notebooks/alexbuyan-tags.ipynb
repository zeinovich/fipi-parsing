```python
import sqlite3
import json
from typing import List

def get_unique_tags(db_path: str) -> List[str]:
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()

    # Get all tags columns
    cursor.execute("SELECT tags FROM tasks")
    all_tags = cursor.fetchall()
    
    # Parse JSON strings and collect unique tags
    unique_tags = set()
    for tags_json, in all_tags:
        if tags_json:
            tags = json.loads(tags_json)[0]
            # print(tags)
            unique_tags.update(tags)
    
    conn.close()
    return sorted(list(unique_tags))
```

```python
DEFAULT_DB_PATH = "../artifacts/tasks.db"

unique_tags = get_unique_tags(DEFAULT_DB_PATH)
unique_high_level_tags = set()
for tag in unique_tags:
    print(tag)
    tag_id = tag.split()[0]
    high_level_tag = tag_id[:-2]
    # print(high_level_tag)
    unique_high_level_tags.add(high_level_tag)
print(f'Total tags: {len(unique_tags)}')
print(f'Total high-level tags: {len(unique_high_level_tags)}')
```
